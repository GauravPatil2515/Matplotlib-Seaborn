```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 100
```

```python
nums = np.random.uniform(0, 10, 100)
plt.bar(np.arange(100),height= nums)
```

```python
plt.barh(np.arange(100),width= nums)
```

```python
plt.bar(np.arange(100),height= nums, color='green')
plt.bar(np.arange(100),height= nums, color='limegreen')
plt.bar(np.arange(100),height= nums, color='lime',edgecolor='black')
```

```python
plt.bar(np.arange(100),nums,align ="center",hatch ="//")
plt.bar(np.arange(100),nums,align ="edge",hatch ="\\")
```

```python
rands =np.random.normal(size=int(1e6))
plt.hist(rands,bins=100)
```

```python
plt.hist(rands,bins=100,cumulative=True)
plt.clf( )
```

```python
plt.hist(rands,range=(0,1),bins=10,cumulative=True)
```

```python
plt.hist(rands,range=(0,6))
```

```python
plt.hist(rands,bins=(1,10),density=True)
```

```python
plt.hist(rands,bins="auto",histtype="stepfilled")
```

```python
plt.hist((rands,0.5*rands),bins="auto",stacked=True)
```
