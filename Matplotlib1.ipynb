```python
import numpy as np
import matplotlib.pyplot as  plt
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 100
```

```python
xvals = np.arange(0, 10, 0.1)
```

```python
plt.plot(xvals, np.sin(xvals))
plt.show()
```

```python
plt.plot(xvals, np.sin(xvals))
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
plt.plot(xvals, np.sin(xvals), label='sin(x)')
plt.plot(xvals, np.cos(xvals), label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.plot(xvals, np.sin(xvals), label='sin(x)',color="red")
plt.plot(xvals, np.cos(xvals), label='cos(x)',color="cyan")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.savefig('plot.png')
plt.legend()
```

```python
plt.plot(xvals, np.sin(xvals), label='sin(x)',color="red",linestyle ='-')
plt.plot(xvals, np.cos(xvals), label='cos(x)', marker ="o")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-1.5,1.5)
plt.xlim(0,10)
plt.legend()
plt.show()
```

```python
rands = np.random.normal(size=1000)

plt.hist(rands)
plt.xlabel('xr')
plt.ylabel('count')
plt.show()
```

```python
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size =(512,512)),sigma=5)
plt.imshow(rands2d)
plt.show()
#
```

```python
plt.imshow(rands2d,cmap="inferno")
plt.colorbar()
plt.show()
#
```
