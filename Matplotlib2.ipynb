```python
import numpy as np
import matplotlib.pyplot as  plt
```

```python
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 100
```

```python
xvals = np.arange(0, 10, 0.1)
```

```python
plt.plot(xvals, np.sin(xvals))
plt.show()
```

```python
idx = 0
for marker in ('.', ',', 'o', 'v', '^', '<', '>', '|', '_'):
  plt.plot(xvals, np.sin(xvals)+idx,''.join(('k',marker)),markersize=10)
  idx += 1
plt.legend()
plt.show()
```

```python
plt.plot(xvals, np.sin(xvals),"k.--",color ="r")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
plt.plot(xvals, np.sin(xvals), label='sin(x)',linestyle ="none",marker =r"$sin$")
plt.plot(xvals, np.cos(xvals), label='cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
plt.plot(xvals, np.sin(xvals), label='sin(x)',color="red",linewidth =10)
plt.plot(xvals, np.cos(xvals), label='cos(x)',color="cyan")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.savefig('plot.png')
plt.legend()
```

```python
plt.plot(xvals, np.sin(xvals), label='sin(x)',color="red",linestyle ='-',alpha =0.5 )
plt.plot(xvals+1, np.cos(xvals), label='cos(x)', marker ="o")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-1.5,1.5)
plt.xlim(0,10)
plt.legend()
plt.show()
```

```python
xvals = np.arange(0, 10, 0.1)
plt.scatter(xvals, np.sin(xvals), label='sin(x)',c=np.cos(xvals),edgecolor = 'black')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
plt.scatter(xvals, np.cos(xvals), label='cos(x)',c=np.cos(xvals),cmap ="inferno",edgecolor = 'black',s =np.power(xvals,3))
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
rands = np.random.normal(size=1000)

plt.hist(rands)
plt.xlabel('xr')
plt.ylabel('count')
plt.show()
```
